**10.** Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
# creating a new column
data_1.dropna(subset=['total_bedrooms'], inplace=True)
data_1['total_bedrooms'] = data_1['total_bedrooms'].astype(int)
```

```python
def map_bedroom_size(total_bedroom):
   if total_bedroom<=10:
      return 'Small'
   elif 10<total_bedroom<=1000 :
     return 'Medium'
   else:
     return 'large'


data_1['total_bedroom_size'] = data_1['total_bedrooms'].map(map_bedroom_size)
print(data_1.head(10))
```
